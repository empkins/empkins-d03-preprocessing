```python
from empkins_io.datasets.d03.micro_gapvii import MicroBaseDataset

import json

from pathlib import Path

import shutil
```

```python
deploy_type = "local"
```

```python
config_dict = json.load(Path("../config.json").open(encoding="utf-8"))

base_path = Path(config_dict[deploy_type]["base_path"])
base_path
```

```python
video_in_path = Path("/Volumes/Untitled/D03Micro_video_backups/face_recordings/mp4")
```

```python
dataset = MicroBaseDataset(base_path)
dataset.get_subset(subject="VP_001", condition="tsst").face_video_path
```

```python
for subset in dataset.groupby(["subject", "condition"]):
    print(subset.subject, subset.condition)

    # get video in path 
    video_in_path_per_subject = video_in_path.joinpath(f"video_face_{subset.subject.lower()}_{subset.condition}.mp4")

    # get video out path
    video_out_path = subset.face_video_path.parent

    print(video_out_path)

    if video_in_path_per_subject.exists() and video_in_path_per_subject.is_file():
        # copy in to out 
        shutil.copy(video_in_path_per_subject, video_out_path)
```

```python
# rename videos

for video in video_in_path.glob("*.mp4"):
    
    split = video.name.split("_")
    video.rename(video_in_path.joinpath(f"video_face_vp_{split[3].zfill(3)}_{split[4]}"))
```
